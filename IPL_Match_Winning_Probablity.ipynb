Every match has a two innings 1st innig and 2nd innings we want to find total runs including both the innings.in every match 

```python
# will do group by to achive this 
total_score_df = delivery.groupby(["match_id","inning"]).sum()["total_runs"].reset_index()
```

```python
total_score_df
```

```python
#now we will keep total_score of first innigs only 
total_score_df = total_score_df[total_score_df["inning"]==1]
```

```python
total_score_df
```

now we will merge the two data frames,so that we will get the match details and total score in single df

```python
match_df = match.merge(total_score_df[["match_id","total_runs"]],left_on="id",right_on="match_id")
```

```python
# now we got the total runs of first innigs along with the match details
match_df
```

```python
match_df["team1"].unique()
```

```python
teams = ["Sunrisers Hyderabad","Mumbai Indians","Royal Challengers Bangalore",
        "Kolkata Knight Riders","Kings XI Punjab","Chennai Super Kings","Rajasthan Royals",
        "Delhi Capitals"]

```

replace the teams with proper name cause they are haveing multiple names for same team

```python
match_df["team1"] = match_df["team1"].str.replace("Delhi Daredevils","Delhi Capitals")
match_df["team2"] = match_df["team2"].str.replace("Delhi Daredevils","Delhi Capitals")
```

```python
match_df["team1"] = match_df["team1"].str.replace("Deccan Chargers","Sunrisers Hyderabad")
match_df["team2"] = match_df["team2"].str.replace("Deccan Chargers","Sunrisers Hyderabad")
```

```python
#now will use isin to keep the teams which is in our define list
match_df = match_df[match_df["team1"].isin(teams)]
match_df = match_df[match_df["team2"].isin(teams)]
```

```python
match_df.shape
```

```python
# will check how many match are impactes due to rain and will romve those matches which are
# impacted due to rain
match_df["dl_applied"].value_counts()
```

```python
match_df = match_df[match_df["dl_applied"]==0]
```

```python
match_df
```

now we will extract those from match_df those we actually reqired

```python
match_df = match_df[["match_id","city","winner","total_runs"]]
```

```python
# now we will merge this match_df with delivery to get ball by ball data
delivey_df = match_df.merge(delivery,on="match_id")
```

```python
delivey_df
```

```python
# now we will keep only those value where innigs is 2 cause we want to predict while chasing
```

```python
delivery_df = delivey_df[delivey_df["inning"] ==2]
```

```python
delivery_df
```

now we will need to find the total score after evry match

```python
delivery_df["current_score"] = delivery_df.groupby("match_id").cumsum()["total_runs_y"]
#cumsum() function is used to find the cumulative sum 
```

```python
delivery_df
```

so now at the end we are having current score and we are already having the total runs so we can find runs left

```python
delivery_df["runs_left"] = delivery_df["total_runs_x"] - delivery_df["current_score"]
```

```python
delivery_df
```

now we will try to find how many balls are left

in data we have given the over and ball of the over so will simply subtract the combination



```python
delivery_df["balls_left"] = 126 - (delivery_df["over"]*6 + delivery_df["ball"])
```

```python
delivery_df
```

now we got the runs left and balls left now will try to find wicket left

```python
delivery_df["player_dismissed"] = delivery_df["player_dismissed"].fillna("0")
delivery_df["player_dismissed"] = delivery_df["player_dismissed"].apply(lambda x:x if x=="0" else 1)
delivery_df["player_dismissed"] = delivery_df["player_dismissed"].astype("int")
wickets = delivery_df.groupby("match_id").cumsum()["player_dismissed"].values
delivery_df["wickets"] = 10 - wickets
```

```python
delivery_df.head()
```

```python
delivery_df.tail()
```

so now we will try to find the cuurent run rate and reqired run rate

```python
# currennt run rate
# crr = runs/overs

delivery_df["crr"] = (delivery_df["current_score"]*6)/(120-delivery_df["balls_left"])
```

```python
delivery_df.head()
```

```python
# now we will calculate the reqired run rate
delivery_df["rrr"] = (delivery_df["runs_left"]*6)/delivery_df["balls_left"]
```

```python
delivery_df
```

now we have done with all the columns only result columns is left

so since we are predicting for second innig so if winner team and batting team is same then our result will be 1 otherwise 0

```python
# will craete one function
def result(row):
    return 1 if row["batting_team"] == row["winner"] else 0
```

```python
delivery_df["result"] = delivery_df.apply(result,axis=1)
```

```python
delivery_df
```

```python
# now we will extract the exact columns which we want to train the model
final_df = delivery_df[["batting_team","bowling_team","city","runs_left","balls_left","wickets","total_runs_x",
            "crr","rrr","result"]]
```

```python
final_df.rename(columns={"wickets":"wickets_left"},inplace=True)
```

```python
final_df.rename(columns={"wickets_left":"wickets"},inplace=True)
```

```python
final_df
```

so now we will shuffule the data becasue it is a match wise data can be biased to the model.

```python
final_df = final_df.sample(final_df.shape[0])
```

```python
final_df
```

```python
final_df.isnull().sum()
```

```python
final_df.dropna(inplace=True)
```

```python
final_df = final_df[final_df["balls_left"] !=0 ]
```

```python
# now we will start the model building
# wii create X and Y
X = final_df.iloc[:,:-1]
Y = final_df.iloc[:,-1]
```

```python
#train test split
from sklearn.model_selection import train_test_split as tts

X_train,X_test,Y_train,Y_test = tts(X,Y,test_size=0.2,random_state=1)
```

```python
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder


trf = ColumnTransformer([("trf",OneHotEncoder(sparse=False,drop="first"),["batting_team",
    "bowling_team","city"])],remainder="passthrough")
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
```

```python
pipe = Pipeline(steps=[("step1",trf),("step2",LogisticRegression(solver="liblinear"))])
```

```python
pipe.fit(X_train,Y_train)
```

```python
Y_pred = pipe.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report

cfm = confusion_matrix(Y_test,Y_pred)
print(cfm)

print("classification report")

print(classification_report(Y_test,Y_pred))

acc = accuracy_score(Y_test,Y_pred)
print("Accuracy of the model:",acc)
```
